```python
# required libraries

import pandas as pd #analysis
import matplotlib.pyplot as plt #vizualization
```

```python
pokemon_df = pd.read_csv("/content/Pokemon.csv")
pokemon_df.head(n=10)
```

```python
pokemon_df.sample(5)
```

```python
# How many ROWS and COLUMNS are in this data?

rows, columns = pokemon_df.shape # == tuple "(800, 13)", saves it to rows and columns, .shape --> returns a tuple that represents the dimensions of the DataFrame
print(f"Number of rows is {rows}, number of columns is {columns}")
```

```python
# How many UNIQUE Pokemon names are in the list?

unique_names = pokemon_df["Name"].nunique() # nunique --> number of unique; unique --> all unique variables
unique_names
```

```python
pokemon_df.head(2)
```

```python
# How many Pokemons have Attack GREATER THAN 100?

attack_greater_than_100 = pokemon_df[pokemon_df['Attack'] > 100]
print(len(attack_greater_than_100))
```

```python
# How many Pokemons have Defese GREATER THAN 100?

defense_greater_than_100 = pokemon_df[pokemon_df["Defense"] > 100]
print(len(defense_greater_than_100))
```

```python
# How many Pokemons have BOTH (Attack and Defense) GREATER THAN 100?
```

```python
attack_defense_greater_than_100 = pokemon_df[(pokemon_df['Attack'] > 100) & (pokemon_df['Defense'] > 100)]
print(len(attack_defense_greater_than_100))
```

```python
# VALUE COUNTS!!!

# Create a grouped DataFrame that COUNTS the number of Pokemons in each group for Type 1.

type1_pokemons = pokemon_df["Type 1"].value_counts().reset_index() # selects column to group by
type1_pokemons

# reset_index() adds index column
```

```python
# BAR CHART

type1_pokemons.plot(x = "Type 1", y = "count", kind = "bar", color = "white", edgecolor = "black")
plt.title("Pokemon count by Type 1")
plt.xlabel("Type 1")
plt.ylabel("Count of Pokemons")

# plot wrapper on dataframe
```

```python
plt.bar(type1_pokemons['Type 1'], type1_pokemons['count'], color=['thistle','skyblue'], edgecolor='maroon')
plt.xticks(rotation=90)
plt.title("Pokemon count by Type 1")
plt.xlabel("Type 1")
plt.ylabel("Count of Pokemons")
```

```python
# Which Pokemon has the BIGGEST attack?

pokemon_df["Attack"].max() #the highest value of the attack
pokemon_df["Attack"].idxmax() #index of the highest value

highest_attack = pokemon_df["Name"].iloc[pokemon_df["Attack"].idxmax()] # iloc --> integer-location based indexing
print("Pokemon with the highest attak is:", highest_attack)

# filter index, and find element by it;
```

```python
# Which Pokemon has the LOWEST attack?

pokemon_df["Attack"].min() #the lowest value of the attack
pokemon_df["Attack"].idxmin() #index of the lowest value

lowest_attack = pokemon_df["Name"].iloc[pokemon_df["Attack"].idxmin()] #iloc --> integer-location based indexing
print("Pokemon with the lowest attak is:", lowest_attack)
```

```python
# ADDING a column

pokemon_df["My unique column"] = "Pokemon"
pokemon_df
```

```python
# ADDING calculated column
pokemon_df["Attack_Defense"] = pokemon_df["Attack"] + pokemon_df["Defense"]
pokemon_df
```

```python
# Quick MATH CALCULATIONS

pokemon_df["Attack"].sum()
pokemon_df["Attack"].min()
pokemon_df["Attack"].max()
pokemon_df["Attack"].mean() # average
```

```python
# Check if value is HIGHER THAN AVERAGE

pokemon_df["Attack"] > pokemon_df["Attack"].mean()

# Add it to df with new column

pokemon_df["Attack above average"] = pokemon_df["Attack"] > pokemon_df["Attack"].mean()

pokemon_df
```

```python
# ADD/INSERT item to column 9

pokemon_df.insert(9, 'Attack_Defense', pokemon_df.pop('Attack_Defense'))
pokemon_df
```

```python
# REMOVE a column

pokemon_df.shape
pokemon_df = pokemon_df.drop(columns=["My unique column"])
pokemon_df.shape
```

# HISTOGRAM


```python
# Analyze "Total" for each Pokemon

plt.hist(pokemon_df["Total"], color = "green", edgecolor = "black", alpha = 0.7)

# aplha --> transparency of color
# default--> 10 bins

plt.xlabel("Total")
plt.ylabel("Number of Pokemons (Frequency)")

# average line

plt.axvline(pokemon_df["Total"].mean(), linestyle="dashdot", color="gray")

# axvline --> average line, added to x-axes, vertical line

# minline
# maxline

plt.axvline(pokemon_df['Total'].min(), color='red', linestyle='dashdot', linewidth=2)
plt.axvline(pokemon_df['Total'].max(), color='blue', linestyle='dashdot', linewidth=2)

#plt.grid(axis = "x", linestyle="--", alpha=0.7)

plt.grid(linestyle="--", alpha=0.7)
```

```python
pokemon_df["Total"].mean()
```

```python
plt.hist(pokemon_df["Total"], bins=3, color = "green", edgecolor = "black", alpha = 0.7)

# (array([276., 411., 113.]), --> How many values are there in bins
# array([180., 380., 580., 780.]), --> start and finish value of the bin
# <BarContainer object of 3 artists>)
```

```python
value = plt.hist(pokemon_df["Total"], bins=3, color = "green", edgecolor = "black", alpha = 0.7)

hist_values=value[0]
bin_edges = value[1]

# accessing element 0 or 1 from the object that is created by hist (e.g value())
# .hist() method returns an object with certain parameters

print('Histogram counts:', hist_values)
print('Bin edges', bin_edges)
```
