````python
# MATPLOTLIB -->

# plotting library in Python
# used for data vizualization
# https://matplotlib.org/stable/gallery/index.html


# sample data: representing the average temperature (in degrees Celsius) from 2010 to 2014
# data = object = dictionary

data = {'Year': [2010, 2011, 2012, 2013, 2014],
        'Avg_Temperature': [14.1, 14.3, 14.5, 14.2, 14.6]}
import pandas as pd
import matplotlib.pyplot as plt # sub-library "pyplot"; abbreviation "plt"
df = pd.DataFrame(data)
df

# VERSION ONE (= simple line chart)

plt.plot(df['Year'], df['Avg_Temperature'], marker="o")

plt.title("Average temperature from 2010 to 2014") # add title
plt.xlabel("Year") # title of x-axis
plt.ylabel("Temperature") # title of y-axis

plt.xticks(df["Year"]) # year values are rounded
plt.grid(True) # add grid
plt.show() # show vizualisation (wonÂ´t print logs of the vizualisation before plot, if used)

# ticks --> values of x -and y-axis
# square brackets [] --> refer to column
# 1st value = X-axis; 2nd value = Y-axes
````

````python
## Grid
plt.grid(True)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.minorticks_on()
````

````python
# VERSION TWO

# difference in code (vs version one) --> 1st line of code
# main difference (vs version one) --> customization of plot; to create/control customizations --> use 1st

# df.plot --> for quick, simple data vizualisations

df.plot(x='Year', y='Avg_Temperature', kind='line') # plotting wrapper, cannot have df with same column names, need to define what kind of chart we use
plt.title('Average temperature from 2010 to 2014') # add title
plt.xlabel('Year') # title of x-axis
plt.ylabel('Temperature') # title of y-axis

plt.xticks(df['Year'])
plt.grid(True)
plt.show() #show visualization

````

````python
# HISTOGRAM - presents the distribution of a single variable; frequency

data2 = {'Scores': [85, 92, 78, 20, 89, 90, 82, 56, 60, 23, 90, 40, 77]} # dictionary, key is "scores"
df2 = pd.DataFrame(data2) # input data2 to DataFrame
df2

plt.hist(df2["Scores"], bins=3, color="green", edgecolor="black")

# plt = print, column "Scores"
# bins = in how many groups data will be grouped, will divide data in equal sizes

# array([20., 44., 68., 92.]) = bins 20-43; 44-67; 68-92
````

````python
# CUSTOM BIN EDGES

bin_edges = [0, 40, 60, 80, 90, 100]
plt.hist(df2["Scores"], bins=bin_edges, color ="red", edgecolor= "black")
````

````python
# SCATTER PLOT - used to visualize the relationship btw two variables

data3 = {"X": [1,2,3,5,5,2], "Y": [5,8,7,4,9,5]} # has to be equal nr of data, otherwise error

df3 = pd.DataFrame(data3)
df3

plt.scatter(x = df3["X"], y =df3["Y"])
plt.title("Scatter Plot")
plt.grid(True)

# Markers:
# https://matplotlib.org/stable/gallery/lines_bars_and_markers/scatter_star_poly.html
````

````python
# Sample (by Laila)

import numpy as np
# Linear trendline
z = np.polyfit(df3['X'], df3['Y'], 1)
p = np.poly1d(z)
# Plot the trendline
plt.plot(df3['X'], p(df3['X']), color='red')
````

````python
# To change grid color

plt.grid(True, which='both', linestyle='--', linewidth=0.1, color='red')
````

````python
# PIE CHART - represents parts of a whole

data4 = {'Category': ['A', 'B', 'C', 'D'], 'Values': [15, 30, 25, 40]}
df4 = pd.DataFrame(data4)
df4
````

````python
colors = ['pink', 'cyan', 'skyblue', 'yellow'] #counterclokwise

plt.pie(df4["Values"], labels = df4["Category"], autopct="%1.1f%%", colors=colors) #floating point at decimetal 1,1
plt.title("Pie Chart")
plt.show()

````

````python
# BOX PLOT - distribution of dataset and identifies outliers

plt.boxplot(df4["Values"])
plt.grid(True)
plt.show()

# in the middle - middle value, then outlayers
# 25% quartiles; 50% within the box
````

````python
# TASK 1:
# Create a bar chart that shows the count of customers for each unique value in the 'Gender' column (including NaN values). - easy
````

````python
import matplotlib.pyplot as plt
import pandas as pd

dataset = pd.read_csv('/content/transaction_dataset.csv')
dataf = pd.DataFrame(dataset)

dataset['Gender'] = dataset['Gender'].astype(str)
dataset['Gender'].value_counts()
gender_counts = dataset['Gender'].value_counts(dropna=False).reset_index()
gender_counts.columns = ['Gender', 'Counts']
df = pd.DataFrame(gender_counts)

plt.bar(df['Gender'], df['Counts'], color=['thistle','skyblue', 'orange'], edgecolor='maroon')

plt.title('Customers per Gender')
plt.xlabel('Gender')
plt.ylabel('Amount')
````

````python
# TASK 2:
# Create a horizontal bar chart that shows the top 5 most frequent names in the DataFrame, based on the 'name' column. (First, create a grouped DataFrame (name_df), then filter it using iloc, and finally create the visualization.) - medium
````

````python
name_df = pd.DataFrame(dataset['Name'].value_counts().reset_index())
name_df.columns = ['Name', 'Counts']

filtered_names = name_df.iloc[:5]

plt.barh(filtered_names['Name'], filtered_names['Counts'], color='skyblue')
plt.title('Top 5 names')
plt.ylabel('Names')
plt.xlabel('Frequency')

plt.show()
````
