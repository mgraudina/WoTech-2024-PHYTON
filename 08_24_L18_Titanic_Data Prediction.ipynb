Data IMPORT

```python
import seaborn as sns
import pandas as pd

titanic_data = sns.load_dataset('titanic')

# seaborn has several datasets, titanic is one of them
```

```python
titanic_data.head()
```

Data INVESTIGATION

```python
titanic_data.isnull().sum() # checking null values
```

```python
titanic_data.dtypes # checking data types
```

```python
# titanic_data.drop('deck', axis=1, inplace=True) # a lot of empty values
```

```python
titanic_data.drop(['alive', 'alone', 'adult_male'], axis=1, inplace=True)
```

```python
titanic_data.drop(['class'], axis=1, inplace=True) # sames as pclass
```

### Data TRANSFORMATION

```python
titanic_data['class'] = titanic_data['class'].astype(str) # modified the object from category > to > string

titanic_data.fillna(0, inplace=True) # replaces empty with null
```

```python
titanic_data.head()
```

```python
titanic_data['sex'].unique()
```

```python
titanic_data['who'].unique()
```

```python
# Sex column

sex_mapping = {'male':0, 'female':1} # dictionary
titanic_data['sex'] = titanic_data['sex'].map(sex_mapping) # replacement in columns that contain object
```

```python
titanic_data.head()
```

```python
# Embarked town

titanic_data['embark_town'].unique() # 0 vales >> cannot be replaced with 0
embarked_town_mapping = {'Southampton':1, 'Cherbourg':2, 'Queenstown':3, 0:0}
titanic_data['embark_town'] = titanic_data['embark_town'].map(embarked_town_mapping)
```

```python
titanic_data.head()
```

```python
# Embarked

titanic_data['embarked'].unique()
embarked_mapping = {'S':1, 'C':2, 'Q':3, 0:0}
titanic_data['embarked'] = titanic_data['embarked'].map(embarked_mapping)
titanic_data.head()
```

```python
# Who

titanic_data['who'].unique()
who_mapping = {'man':1, 'woman':2, 'child':3}
titanic_data['who'] = titanic_data['who'].map(who_mapping)
titanic_data.head()
```

```python
titanic_data.info() # now all data --> int or float
```

```python
# LabelEncoder > labels values 0 to n-1 (to replace all in one go)
# https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html
```

DATA MODELING

```python
# Trying to predict who would survive in Titanic Disaster

from sklearn.model_selection import train_test_split

X = titanic_data.drop('survived', axis=1) # FEATURES: exluding dependent variable survived, column (=1)
y = titanic_data['survived'] # DEPENDENT VARIABLE: only keeping survived column
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)

# random state = every time you run the code, the split is the same, not a rule
```

Logistic regression

```python
from sklearn.linear_model import LogisticRegression # import
```

```python
model = LogisticRegression() # initialize
```

```python
model.fit(X_train, y_train) # fit
```

```python
predicition_results = model.predict(X_test)
predicition_results

# probablity: 0 - did not survive; 1 - survived
```

```python
from sklearn.datasets import load_iris
from sklearn import tree
iris = load_iris()
X_train, y_train = iris.data, iris.target
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)

tree.plot_tree(clf)
```

**Cross-validation**

```python
import numpy as np
from sklearn.model_selection import KFold, cross_val_score

k_fold = KFold(n_splits=10, shuffle=True, random_state=42)
for train, test in k_fold.split(X_train):
    print("%s %s" % (train, test))
accuracy_scores = cross_val_score(model_tree, X, y, cv=k_fold, scoring='accuracy')
print(f'The accuracy scores are {accuracy_scores}')
```
