#### RANDOM WALK

```python
# watch video from Datiorium on 'random walk'
# mathematical concept
```

```python
n = 100000 # defining number of steps

x = np.zeros(n)
y = np.zeros(n)
```

```python
x
```

```python
import random
val = random.randint(1,4) # random int btw 1-4 = direction (right, left, up, down)
val
```

```python
for i in range(1, n):
  val = random.randint(1, 4)

  if val == 1:
    x[i] = x[i - 1] + 1 # dot moves x coordinate +1, horizontal, moves to the right
    y[i] = y[i - 1] # stays the same, as previous
  elif val == 2:
    x[i] = x[i - 1] - 1 # dot moves x coordinate -1, horizontal, moves to the left
    y[i] = y[i - 1]
  elif val == 3:
    x[i] = x[i - 1]
    y[i] = y[i - 1] + 1 # dot moves y coordinate +1
  else: # val == 4:
    x[i] = x[i - 1]
    y[i] = y[i - 1] - 1 # dot moves y coordinate -1

  # our line starts from previous position [i-1]
```

```python
import pylab
```

```python
pylab.title(f'Random walk of n=100000 steps')
pylab.plot(x, y)
```

```python
# shows start and end (code by MMari)

plt.figure(figsize=(10, 6))  # Increase figure size
plt.plot(x, y, color='blue', lw=2, alpha=0.7)  # Line with transparency
plt.scatter(x[0], y[0], c='green', s=100, label='Start')  # Mark start point
plt.scatter(x[-1], y[-1], c='red', s=100, label='End')  # Mark end point

plt.title('2D Random Walk', fontsize=16)
plt.xlabel('X-Coordinate')
plt.ylabel('Y-Coordinate')
plt.grid(True)
plt.legend()

plt.show()
```

```python
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.savefig.html

# plt.savefig('2D_rand') # png by default
```

```python
# all steps in different color

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
steps = 100
x = np.cumsum(np.random.choice([-1, 1], size=steps))
y = np.cumsum(np.random.choice([-1, 1], size=steps))

plt.figure(figsize=(10, 6))

# creates colors
colors = plt.cm.viridis(np.linspace(0, 1, len(x) - 1))

for i in range(len(x) - 1):
    plt.plot(x[i:i+2], y[i:i+2], color=colors[i], lw=2, alpha=0.7)

plt.scatter(x[0], y[0], c='green', s=100, label='Start')
plt.scatter(x[-1], y[-1], c='red', s=100, label='End')

plt.title('2D Random Walk', fontsize=16)
plt.xlabel('X-Coordinate')
plt.ylabel('Y-Coordinate')
plt.grid(True)
plt.legend()

plt.show()
```

####TEAMWORK

Simulate a 1D simple random walk where at each step can be moving only up or down. How many steps   can you actually get?

```python
import random
import matplotlib.pyplot as plt

n = 1000
position = [0]
steps = 0

for i in range(1, n):
    step = random.choice([-1, 1])
    position.append(position[-1] + step)

    if step == 1:
        steps += 1

plt.figure(figsize=(10, 6))
plt.plot(position, label='Position Over Time')
plt.title(f'1D Simple Random Walk - Total Up Steps: {steps}', fontsize=16)
plt.xlabel('Step')
plt.ylabel('Position')
plt.grid(True)
plt.legend()
plt.show()

print(f"Total number of up steps: {steps}")
```
